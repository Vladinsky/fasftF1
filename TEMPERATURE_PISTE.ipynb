```python
import matplotlib.pyplot as plt
import fastf1
import fastf1.plotting
```

```python


# Carica i dati della sessione (es: Gran Premio, anno, sessione)
session = fastf1.get_session(2023, 'Monza', 'R')  # Sostituisci con l'anno, il circuito e la sessione desiderati
session.load()  # Carica i dati

# Ottieni i dati meteo
weather_data = session.weather_data

# Plot della temperatura del tracciato
plt.figure(figsize=(10, 5))
plt.plot(weather_data['Time'], weather_data['TrackTemp'], label='Temperatura del tracciato (C)', color='red')
plt.title(f"Temperatura del tracciato - {session.event['EventName']} {session.name} {session.event.year}")
plt.xlabel('Tempo')
plt.ylabel('Temperatura (C)')
plt.legend()
plt.grid()
plt.show()
```

```python
import fastf1 as ff1
import matplotlib.pyplot as plt


# Anno di interesse
year = 2023  # Puoi cambiare l'anno se necessario

# Lista di tutti i circuiti della stagione
events = ff1.get_event_schedule(year)

# Filtra solo gli eventi di Formula 1 (escludendo altre serie)
events = events[events['EventFormat'] == 'conventional']

# Crea un grafico per ogni circuito
for _, event in events.iterrows():
    try:
        # Carica la sessione di gara (puoi cambiare 'R' con 'Q' per le qualifiche)
        session = ff1.get_session(year, event['EventName'], 'R')
        session.load()

        # Ottieni i dati meteo
        weather_data = session.weather_data

        # Plot della temperatura del tracciato
        plt.figure(figsize=(10, 5))
        plt.plot(weather_data['Time'], weather_data['TrackTemp'], label='Temperatura del tracciato (C)', color='red')
        plt.title(f"Temperatura del tracciato - {event['EventName']} {year}")
        plt.xlabel('Tempo')
        plt.ylabel('Temperatura (C)')
        plt.legend()
        plt.grid()
        plt.show()

    except Exception as e:
        print(f"Errore durante il caricamento dei dati per {event['EventName']}: {e}")
```
