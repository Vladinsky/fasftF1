```python

# Carica i dati della sessione
session = fastf1.get_session(2023, 'Monza', 'R')  # Anno, Circuito, Tipo di sessione (R = Gara)
session.load()

# Seleziona il pilota (es. Lewis Hamilton, vettura 44)
driver_code = 'HAM'  # Codice pilota (es. HAM per Hamilton)
driver_laps = session.laps.pick_driver(driver_code)

# Seleziona un giro specifico (es. il giro più veloce)
fastest_lap = driver_laps.pick_fastest()
telemetry = fastest_lap.get_telemetry()

# Estrai i dati di velocità e RPM
speed = telemetry['Speed']  # Velocità in km/h
rpm = telemetry['RPM']  # Giri del motore
distance = telemetry['Distance']  # Distanza percorsa durante il giro

# Crea il grafico con sfondo nero
plt.style.use('dark_background')  # Imposta lo sfondo nero
fig, ax1 = plt.subplots(figsize=(12, 6))

# Grafico della velocità
color = 'red'
ax1.set_xlabel('Distanza (m)')
ax1.set_ylabel('Velocità (km/h)', color=color)
ax1.plot(distance, speed, color=color, label='Velocità')
ax1.tick_params(axis='y', labelcolor=color)

# Grafico dei RPM (sull'asse secondario)
ax2 = ax1.twinx()
color = 'white'
ax2.set_ylabel('RPM', color=color)
ax2.plot(distance, rpm, color=color, label='RPM')
ax2.tick_params(axis='y', labelcolor=color)

# Aggiungi titolo e legenda
plt.title(f"Analisi delle Prestazioni - {driver_code} (Giro {fastest_lap['LapNumber']})", pad=20)
fig.legend(loc='upper right', bbox_to_anchor=(0.9, 0.9))

# Mostra il grafico
plt.tight_layout()
plt.show()
```
